```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot
from datetime import datetime
```

```python
df = pd.read_csv("datasets/energy_dataset.csv",
                sep = ',')
```

```python
df.info()
```

```python
df.drop(columns=['generation hydro pumped storage aggregated','forecast wind offshore eday ahead'], inplace=True)
```

```python
df.head(3)
```

```python
df['date'] = pd.to_datetime(df['time'], utc=True)
#df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df['month'] = df['date'].dt.strftime('%m')
df['year'] = df['date'].dt.strftime('%Y').astype(int)
df['day'] = df['date'].dt.strftime('%d')
df['day_of_year'] = df['date'].dt.dayofyear
```

```python
df_agg = df.groupby(['month','year','day','day_of_year'], as_index=False).sum()
```

```python
df_agg.head(3)
```

```python
df_agg[(df_agg['month'].astype(int)==2) & (df_agg['day'].astype(int)>27)]
```

```python
sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(15,8)})
sns.lineplot(x=df_agg['day_of_year'], y=df_agg['generation solar'], hue = df_agg['year'], palette = "tab10", linewidth=2.5,)
```
